Q01

Converter os seguintes dados do dicionário em um objeto de dataframe usando Pandas. (O objeto dataframe é nomeado como df.)


```
d = {'col1': [1, 2], 'col2': [3, 4], 'col3': [5, 6], 'col4': [7, 8]}
```



````python
# Importa a biblioteca pandas e a renomeia como 'pd' para facilitar o uso
import pandas as pd

# Cria um dicionário com quatro colunas, onde cada chave é o nome da coluna e os valores são listas de dados
d = {'col1': [1, 2], 'col2': [3, 4], 'col3': [5, 6], 'col4': [7, 8]}

# Cria um DataFrame a partir do dicionário 'd', onde as chaves do dicionário se tornam os nomes das colunas no DataFrame
df = pd.DataFrame(d)

# Exibe o DataFrame gerado, que terá quatro colunas e duas linhas de dados
df
````

Q02

No quadro de dados criado na pergunta 1, criar um novo quadro de dados que consiste apenas de dados de coluna com o nome da coluna "col4". (O quadro de dados é nomeado como new_df.)

````python
# Importa a biblioteca pandas e a renomeia como 'pd' para facilitar o uso
import pandas as pd

# Cria um novo DataFrame 'new_df' a partir dos dados contidos na coluna 'col4' do dicionário 'd'
# Aqui, 'data = d['col4']' seleciona apenas os valores da coluna 'col4' para o novo DataFrame
new_df = pd.DataFrame(data=d['col4'])

# Define o nome da coluna do DataFrame 'new_df' como 'col4'
new_df.columns = ['col4']

# Exibe o novo DataFrame, que contém apenas uma coluna chamada 'col4'
new_df
````
